```python
import yfinance as yf
```

```python
#Import AMD
```

```python
amd = yf.Ticker("AMD")
```

```python
amd = amd.history(period="max")
```

```python
amd
```

```python
amd.index
```

```python
#Data Cleaning and Visualization
```

```python
amd.plot.line(y="Close", use_index=True)
```

```python
del amd["Dividends"]
del amd["Stock Splits"]
```

```python
#Set up ML target
```

```python
amd["Tomorrow"] = amd["Close"].shift(-1)
```

```python
amd
```

```python
amd["Target"] = (amd["Tomorrow"] > amd ["Close"]).astype(int)
```

```python
amd
```

```python
amd = amd.loc["2010-01-01":].copy()
```

```python
amd
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=250, min_samples_split=100, random_state=1)

train = amd.iloc[:-100]
test = amd.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
import pandas as pd

preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```
